Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

1. One Sample T-test

```python
m = [8.8, 6.6, 9.5, 11.2, 10.2, 7.4, 8.0, 9.6, 9.9, 9.0, 7.6, 7.4, 10.4, 11.1, 8.5, 10.0, 11.6, 10.7, 10.3, 7.0]

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(m, 10)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

Observation: If the p-value is less than the alpha level of 0.05, we reject the null hypothesis and conclude that the average mass is significantly different from 10 g.

2. Independent (unpaired)two-sample t-test

```python
x1 = [10.8, 10.0, 8.2, 9.9, 11.6, 10.1, 11.3, 10.3, 10.7, 9.7, 7.8, 9.6, 9.7, 11.6, 10.3, 9.8, 12.3, 11.0, 10.4, 10.4]
x2 = [7.8, 7.5, 9.5, 11.7, 8.1, 8.8, 8.8, 7.7, 9.7, 7.0, 9.0, 9.7, 11.3, 8.7, 8.8, 10.9, 10.3, 9.6, 8.4, 6.6, 7.2, 7.6, 11.5, 6.6, 8.6, 10.5, 8.4, 8.5, 10.2, 9.2]

# Perform independent two-sample t-test
t_stat, p_value = stats.ttest_ind(x1, x2, equal_var=True)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

Observation: If the p-value is less than 0.05, we conclude that there is a significant difference between the masses of acorns from upwind and downwind oak trees.

3. ANOVA Test

```python
A = [51, 45, 33, 45, 67]
B = [23, 43, 23, 43, 45]
C = [56, 76, 74, 87, 56]

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(A, B, C)

print("F-statistic:", f_stat)
print("p-value:", p_value)
```

Observation: If the p-value is less than 0.05, we reject the null hypothesis and conclude that there is a significant difference in the mean marks of students in the three sections.

Graphs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. One Sample T-test
m = [8.8, 6.6, 9.5, 11.2, 10.2, 7.4, 8.0, 9.6, 9.9, 9.0, 7.6, 7.4, 10.4, 11.1, 8.5, 10.0, 11.6, 10.7, 10.3, 7.0]
t_stat, p_value = stats.ttest_1samp(m, 10)

plt.figure(figsize=(8, 6))
plt.hist(m, bins=5, edgecolor='black')
plt.axvline(10, color='red', linestyle='dashed', linewidth=2, label='Population Mean (10)')
plt.xlabel('Mass (g)')
plt.ylabel('Frequency')
plt.title('One-Sample T-Test')
plt.legend()
plt.show()

print("One-Sample T-Test:")
print("t-statistic:", t_stat)
print("p-value:", p_value)


# 2. Independent (unpaired)two-sample t-test
x1 = [10.8, 10.0, 8.2, 9.9, 11.6, 10.1, 11.3, 10.3, 10.7, 9.7, 7.8, 9.6, 9.7, 11.6, 10.3, 9.8, 12.3, 11.0, 10.4, 10.4]
x2 = [7.8, 7.5, 9.5, 11.7, 8.1, 8.8, 8.8, 7.7, 9.7, 7.0, 9.0, 9.7, 11.3, 8.7, 8.8, 10.9, 10.3, 9.6, 8.4, 6.6, 7.2, 7.6, 11.5, 6.6, 8.6, 10.5, 8.4, 8.5, 10.2, 9.2]
t_stat, p_value = stats.ttest_ind(x1, x2, equal_var=True)


plt.figure(figsize=(8, 6))
plt.hist(x1, bins=5, alpha=0.5, label='x1')
plt.hist(x2, bins=5, alpha=0.5, label='x2')
plt.xlabel('Mass')
plt.ylabel('Frequency')
plt.title('Independent Two-Sample T-Test')
plt.legend()
plt.show()


print("\nIndependent Two-Sample T-Test:")
print("t-statistic:", t_stat)
print("p-value:", p_value)

# 3. ANOVA Test
A = [51, 45, 33, 45, 67]
B = [23, 43, 23, 43, 45]
C = [56, 76, 74, 87, 56]
f_stat, p_value = stats.f_oneway(A, B, C)

data = [A, B, C]
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['A', 'B', 'C'])
plt.xlabel('Sections')
plt.ylabel('Marks')
plt.title('ANOVA Test')
plt.show()

print("\nANOVA Test:")
print("F-statistic:", f_stat)
print("p-value:", p_value)
```
