## Continuous Uniform Distribution

```python
# prompt: IT industry records the amount of time a software engineer needs to fix a bug in the initial phase of software development in 'debugging.csv'.
# Time needed to fix bugs X is a continuous random variable. Let's see the distribution of X and answer the below questions. 1. Find the probability that a randomly selected software debugging requires less than three hours 2. Find the probability that a rand

# Assuming the time needed to fix a bug follows a continuous uniform distribution
# Let's assume the minimum time is 0 hours and the maximum time is 8 hours (adjust as needed based on your data)
a = 0  # Minimum time
b = 8  # Maximum time

# 1. Find the probability that a randomly selected software debugging requires less than three hours
probability_less_than_3 = uniform.cdf(3, loc=a, scale=b-a)
print("The probability that a randomly selected software debugging requires less than three hours is:", probability_less_than_3)


# 2. Find the probability that a randomly selected software debugging requires more than 5 hours
probability_more_than_5 = 1 - uniform.cdf(5, loc=a, scale=b-a)
print("The probability that a randomly selected software debugging requires more than 5 hours is:", probability_more_than_5)

# 3. Find the probability that a randomly selected software debugging requires between 2 and 6 hours
probability_between_2_and_6 = uniform.cdf(6, loc=a, scale=b-a) - uniform.cdf(2, loc=a, scale=b-a)
print("The probability that a randomly selected software debugging requires between 2 and 6 hours is:", probability_between_2_and_6)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

# Load the data
df_sat_score = pd.read_csv('/content/20sMGjG0yMJtVQGi (1).csv')  # Replace with the actual file path

# **Assuming the SAT scores are in a column named 'Time Taken to fix the bug', replace 'score' with the actual column name**
# Calculate the mean and standard deviation of SAT scores
mean_sat_score = df_sat_score['Time Taken to fix the bug'].mean()
std_sat_score = df_sat_score['Time Taken to fix the bug'].std()

# 1. Probability of scoring less than 800
probability_less_than_800 = norm.cdf(800, loc=mean_sat_score, scale=std_sat_score)
print("The probability that a student will score less than 800 in the SAT exam is:", probability_less_than_800)

# 2. Probability of scoring more than 1300
probability_more_than_1300 = 1 - norm.cdf(1300, loc=mean_sat_score, scale=std_sat_score)
print("The probability that a student will score more than 1300 in the SAT exam is:", probability_more_than_1300)
```
