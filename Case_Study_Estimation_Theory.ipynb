Q1. Sampling Distribution

```python
import scipy.stats as stats

# Given values
mu = 60  # population mean
sigma = 6  # population standard deviation
n = 40  # sample size
sample_mean = 58  # sample mean

# Standard error
SE = sigma / (n**0.5)

# Z-score calculation
z = (sample_mean - mu) / SE

# Probability
probability = stats.norm.cdf(z)
probability
```

We want to calculate a 95% confidence interval for the mean usage.

```python
import numpy as np

# Given values
x_bar = 310
sigma = 89
n = 40
confidence_level = 0.95

# Z-score for 95% confidence
z_star = stats.norm.ppf((1 + confidence_level) / 2)

# Margin of error
ME = z_star * (sigma / np.sqrt(n))

# Confidence interval
CI_lower = x_bar - ME
CI_upper = x_bar + ME
CI_lower, CI_upper
```

Q3. Hypothesis Testing
We want to test whether the mean waiting time to place an order has changed from the previous population mean value of 4.5 minutes.

Null hypothesis (

```python
import scipy.stats as stats
import numpy as np

# Given values for hypothesis testing
population_mean = 4.5  # Population mean under the null hypothesis
sample_mean = 4.8  # Sample mean
sample_std_dev = 0.5 # Sample standard deviation
sample_size = 50 # Sample size
alpha = 0.05 # Significance level

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# Degrees of freedom
df = sample_size - 1

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

# Perform the hypothesis test
if p_value < alpha:
    print(f"Reject the null hypothesis (p-value: {p_value:.3f} < alpha: {alpha})")
else:
    print(f"Fail to reject the null hypothesis (p-value: {p_value:.3f} >= alpha: {alpha})")
```

Q4. P-value Calculation
To find the p-value for a two-tailed z-test where the computed z-statistic is +2.00:

```python
z_stat = 2.00
p_value = 2 * (1 - stats.norm.cdf(z_stat))
p_value
```

We need to test if the mean income is greater than 29,000 rupees. This can be a one-tailed test.

```python
# Given values
x_bar = 30000
mu_0 = 29000
sigma = 8000
n = 400

# Test statistic
z = (x_bar - mu_0) / (sigma / np.sqrt(n))

# P-value for one-tailed test
p_value = 1 - stats.norm.cdf(z)

z, p_value
```

Conclusion: If the p-value is less than 0.05, we reject the null hypothesis and conclude that the mean income is significantly greater than 29,000 rupees, suggesting that the product line could be profitable in the new market.
